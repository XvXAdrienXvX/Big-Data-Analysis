### Structured Data  
- Conforms to a **data model or schema** and is stored in **tabular form** (e.g., relational databases).  
- Commonly generated by enterprise systems like **ERP** and **CRM**.  
- Easy to process and store due to widespread tool support.  
- Examples: Banking transactions, invoices, customer records.  

### Unstructured Data  
- Does not conform to a data model or schema; makes up **80% of enterprise data**.  
- Includes **textual** (e.g., tweets, blog posts) and **binary** (e.g., images, audio, video) files.  
- Requires **special-purpose logic** for processing (e.g., codecs for video files).  
- Cannot be directly queried with SQL; often stored as **BLOBs** in relational databases or in **NoSQL databases**.  
- Grows faster than structured data.  

### Semi-structured Data  
- Has some **structure** but is not relational; often **hierarchical** or **graph-based**.  
- Commonly stored in **text-based formats** like XML and JSON.  
- Easier to process than unstructured data but may require **pre-processing** (e.g., XML validation).  
- Examples: EDI files, spreadsheets, RSS feeds, sensor data.
