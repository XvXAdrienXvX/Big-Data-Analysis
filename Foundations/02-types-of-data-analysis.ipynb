There are four general categories of analytics that are distinguished by the results they produce:

- descriptive analytics
- diagnostic analytics
- predictive analytics
- prescriptive analytics



### Descriptive Analytics

- **Descriptive analytics** focuses on analyzing past events to provide insights and context from historical data.  
- It answers questions about what has already happened, such as:  
  - Sales performance over a specific period (e.g., past 12 months).  
  - Volume of support calls, categorized by severity and geographic location.  
  - Monthly commissions earned by individual sales agents.  
- Approximately **80% of analytics efforts** are descriptive, offering the least strategic value but requiring basic analytical skills.  
- Tools like **ad-hoc reports** and **dashboards** are commonly used to present descriptive analytics.  
- Reports are typically **static**, displaying historical data through data grids, charts, or visualizations.  
- Data is often sourced from operational systems like **Customer Relationship Management (CRM)** or **Enterprise Resource Planning (ERP)** systems.  



### Diagnostic Analytics
- **Diagnostic analytics** focuses on identifying the causes of past events by analyzing data to understand why something happened.  
- It answers questions aimed at uncovering the reasons behind specific phenomena, such as:  
  - Why were Q2 sales lower than Q1 sales?  
  - Why were there more support calls from the Eastern region compared to the Western region?  
  - Why did patient re-admission rates increase over the past three months?  
- Diagnostic analytics provides **greater value** than descriptive analytics but requires a more **advanced skillset**.  
- It often involves collecting data from **multiple sources** and organizing it in a way that supports **drill-down** and **roll-up analysis**.  
- The goal is to uncover relationships and patterns in the data to explain past events.



### Predictive Analytics

- **Predictive analytics** aims to forecast future outcomes based on historical and current data.  
- It uses patterns, trends, and relationships in data to create models for making predictions.  
- Key characteristics:  
  - Focuses on **what might happen** in the future.  
  - Relies on the strength and magnitude of data associations to build predictive models.  
  - Models depend on the conditions of past events; changes in conditions require model updates.  
- Example questions:  
  - What is the likelihood of a customer defaulting on a loan after missing a payment?  
  - How will patient survival rates differ if Drug B is used instead of Drug A?  
  - What is the probability of a customer purchasing Product C after buying Products A and B?  
- Predictive analytics helps identify **risks** and **opportunities** by analyzing data trends and exceptions.  
- It provides higher value than descriptive and diagnostic analytics but requires advanced analytical techniques.  



### Prescriptive Analytics

- **Prescriptive analytics** builds on predictive analytics to recommend specific actions to take.  
- It focuses on determining the **best course of action** and explains why it is the optimal choice.  
- Combines data analysis with situational understanding to provide actionable insights for gaining advantages or mitigating risks.  
- Example questions:  
  - Which of the three drugs delivers the best results for a specific condition?  
  - What is the optimal time to trade a particular stock?  
- Provides the **highest value** among all analytics types but requires:  
  - Advanced skillsets.  
  - Specialized software and tools.  
- Calculates multiple outcomes and suggests the best action for each scenario.  
- Shifts from **explanatory** to **advisory**, often involving simulations of various scenarios to evaluate potential outcomes.
